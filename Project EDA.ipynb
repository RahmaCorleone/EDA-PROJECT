```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('listings.csv')
df.head(10)
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
df.isnull().sum()
```

```python
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
df.dropna(subset=['description', 'picture_url'], inplace=True)
df['neighborhood_overview'].fillna('Unknown', inplace=True)
df['host_location'].fillna('Unknown', inplace=True)
df['host_about'].fillna('Unknown', inplace=True)
df['host_response_time'].fillna(df['host_response_time'].mode()[0], inplace=True)
df['host_response_rate'].fillna(df['host_response_rate'].mode()[0], inplace=True)
df.isnull().sum()
```

```python
df['host_acceptance_rate'].fillna(df['host_acceptance_rate'].mode()[0], inplace=True)
df['host_is_superhost'].fillna('f', inplace=True)
df.isnull().sum()
```

```python
df['host_neighbourhood'].fillna('Unknown', inplace=True)
df['neighbourhood'].fillna('Unknown', inplace=True)
df.isnull().sum()
```

```python
df2=df.head(30)
df2
```

```python
plt.figure(figsize=(20, 16))
sns.histplot(df2['price'], kde=True,bins=20)
plt.title('Distribution of Prices')
plt.show()
```

```python
plt.figure(figsize=(20, 16))
sns.histplot(df2['number_of_reviews'], kde=True,bins=30)
plt.title('Distribution of Number of Reviews')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df2['number_of_reviews'], df2['price'])
plt.title('Price vs Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Price')
plt.show()
```

```python
df2['price'] = df2['price'].replace('[\$,]', '', regex=True).astype(float)
plt.figure(figsize=(10, 6))
sns.boxplot(x='room_type', y='price', data=df2)
plt.title('Price vs Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.show()
```
